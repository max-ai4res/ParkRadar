# Sigmoid logistic function vs weights & input values
- simple study


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import expit

lowRange = [1/(10**x) for x in range (1, 6)]
midRange = [x/10 for x in range(2, 10)]
weigths = lowRange[::-1] + midRange
weights = np.array(weigths)

s = expit(weights * 0.01)
print(weights, s)
plt.plot(s)
plt.show()


s = expit(weights * 0.99)
print(weights, s)
plt.plot(s)
plt.show()
```
